### 연산자
#### 1. 산술 연산자

```python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a ** b)
print(a % b)
```

#### 2. 비교 연산자
- \>   크다(초과)
- <   작다(미만)
- \>=  크거나같다(이상)
- <=  작거나같다(이하)
- ==  같다
- !=  같지않다

```python
print(a > b) # True
print(a < b) # False
print(a >= b) # True
print(a <= b) # False
print(a == b) # False
print(a != b) # True
```

#### 3. 논리연산자
- and(논리곱)
- or(논리합)
- not(논리부정)

```python
# and는 모든 조건이 참일때만 True
print((10 > 5) and (3 < 1) and (5 > 1)) # False
# or는 모든 조건이 거짓일때만 False
# and가 or보다 우선순위가 높다.
print((10 > 5) or (3 < 1) and (4 < 1)) # True
print(not (10 < 5)) # True
```

#### 4. 대입 연산자
=

```python
# 5를 x에 대입
# 오른쪽 값을 왼쪽에 대입
# 변수: x, a, b
x = 5
# 5 = x -> SyntaxError
# b의 값을 a에 대입
a = b
```

#### 5. 복합 대입 연산자
- +=
- -=
- *=
- /=
- %=
- //=
- **=

```python
x = 5
x += 1      # x = x + 1
# 파이썬은 ++, --는 없다.
print(x)
x -= 3      # x = x - 3
print(x)
x *= 3      # x = x * 3
print(x)
x /= 3      # x = x / 3
print(x)
x //= 3     # x = x // 3
print(x)
x %= 3      # x = x % 3
print(x)
x **= 3     # x = x ** 3
print(x)
```

#### 6. 멤버십 연산자
- in: 값이 안에 있으면 True
- not in: 값이 안에 없으면 True

```python
fruits = ["사과", "바나나", "체리"] # 리스트
print("바나나" in fruits) # True
print("오렌지" in fruits) # False
```
